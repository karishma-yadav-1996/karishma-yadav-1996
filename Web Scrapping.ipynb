# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of 
# release) and make data frame. 

```python
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
  
```

```python
movies_name=soup.find_all('td',class_="titleColumn")
movies_name
```

```python
movie_name=[]
for i in movies_name:
            movie_name.append(i.text.replace("\n",""))
movie_name[:100]
```

```python
movies_rating=soup.find_all('td',class_="ratingColumn imdbRating")
movies_rating
```

```python
movie_ratings=[]
for i in movies_rating:
            movie_ratings.append(i.text.replace("\n",""))
movie_ratings[:100]
```

```python
df=pd.DataFrame()
```

```python
df['Movie  name with Year']=movie_name
df['Movie rating']=movie_ratings
df[:100]
```
